```python
def insert(lst, val, index):
    return lst[:index] + [val] + lst[index:]
```

```python
def sum_square(n):
    return sum([i**2 for i in range(1, n+1)])
```

```python
def digitize(n):
    return [int(i) for i in str(n)]
```

```python
def filter_list(lst, condition):
    return [i for i in lst if condition(i)]
```

```python
def multiples_of_3_or_5(n):
    return sum([i for i in range(n) if i % 3 == 0 or i % 5 == 0])
```

```python
def count_positives_sum_negatives(lst):
    count_positives = len([i for i in lst if i > 0])
    sum_negatives = sum([i for i in lst if i < 0])
    return [count_positives, sum_negatives]
```

```python
def categorize_members(data, condition_func):
    return [condition_func(member) for member in data]
```

```python
def queue_time(lst, n):
    totals = [0] * n
    for i in lst:
        idx = totals.index(min(totals))
        totals[idx] += i
    return max(totals)
```
